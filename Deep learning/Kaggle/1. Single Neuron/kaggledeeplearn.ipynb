```python
model = keras.Sequential([
    layers.Dense(1, input_shape=[1]),
])

x = tf.linspace(-1.0, 1.0, 100)
y = model.predict(x)

plt2.figure(dpi=100)
plt2.plot(x, y, 'k')
plt2.xlim(-1, 1)
plt2.ylim(-1, 1)
plt2.xlabel("Input: x")
plt2.ylabel("Target y")
w, b = model.weights # you could also use model.get_weights() here
plt2.title("Weight: {:0.2f}\nBias: {:0.2f}".format(w[0][0], b[0]))
plt2.show()
```
